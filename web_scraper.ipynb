```python
dataframe = pd.DataFrame(columns = df_cols)
season_types = ['Playoffs', 'Regular%20Season']
years = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24']
import time
begin_loop = time.time()

for y in years:
    for s in season_types:
        api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season='+y+'&SeasonType='+s+'&StatCategory=PTS'
        r = requests.get(url=api_url).json()
        tdataframe1 = pd.DataFrame(r['resultSet']['rowSet'], columns = th)
        tdataframe2 = pd.DataFrame({'Year':[y for i in range(len(tdataframe1))],
                                    'Season_type':[s for i in range(len(tdataframe1))]})
        tdataframe3 = pd.concat([tdataframe2,tdataframe1], axis=1)
        if not tdataframe3.empty:
            dataframe = pd.concat([dataframe, tdataframe3], axis=0)
        print(f'Finished scraping data for the {y} {s}. ')
        lag = np.random.uniform(low = 5, high = 40)
        print(f'...waiting {round(lag, 1)} seconds.')
        time.sleep(lag)
    print(f'Process completed! Total run time: {round((time.time()-begin_loop)/60,2)}')
    dataframe.to_excel('nba_player_data.xlsx', index =False, engine='openpyxl')
```
